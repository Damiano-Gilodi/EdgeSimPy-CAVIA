```python
import pandas as pd

df = pd.read_pickle("processed_data/service_cavia_mean.pkl")
df
```

```python
print(f"Total rows: {len(df)}")
print(f"Distributions: {df['Distribution'].nunique()}")
print(f"Scenarios: {df['Scenario'].nunique()}")
print(f"App: {df[['Scenario', 'App_ms']].drop_duplicates().shape[0]}")
print(f"Average percentage error: {df['Rel_Error_Percent'].mean():.2f}")
print(f"Maximum percentage error: {df['Rel_Error_Percent'].max():.2f}")
print(f"Average percentage error per distribution:\n {df.groupby('Distribution')['Rel_Error_Percent'].mean().round(2)}")
```
